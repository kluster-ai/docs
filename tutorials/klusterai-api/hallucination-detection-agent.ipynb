## Introduction

Hallucinations in AI occur when models generate information that appears plausible but is factually incorrect or unsupported by the provided context. This poses significant risks in production applications, particularly in domains where accuracy is critical.

This tutorial demonstrates how to use <a href="/get-started/hallucination-agent/starting-guide/hallucination-detection-agent/" target="_blank">kluster.ai Hallucination Detection Agent</a> to identify and prevent hallucinations in your applications. We'll explore both detection methods: Question/Answer validation and Chat Completion analysis.

The agent can evaluate AI responses based on provided context (perfect for RAG applications) or perform real-time fact-checking against general knowledge. By following this tutorial, you'll learn how to:

- Detect hallucinations in individual Q&A pairs.
- Compare fact-checking vs. context validation modes.
- Validate responses in full conversation histories.

### Q/A Features

The Question/Answer method validates whether an answer to a specific question contains hallucinated information. It operates in two modes:

1. **Fact-checking mode** - when no context is provided, the agent verifies answers by comparing it to other sources.
2. **Context validation mode** - when context is provided, the agent only validates answers against that context.

Let's start with fact-checking examples using the dedicated endpoint:

```python
# Function that runs the detection for fact-checking examples 
def detect_hallucination_qa(prompt, output, context=None, return_search_results=False):
    """Detect hallucination using the dedicated endpoint"""
    url = base_url_endpoint
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    
    # Prepare the payload
    payload = {
        "prompt": prompt,
        "output": output,
        "return_search_results": return_search_results
    }
    
    # Add context if provided
    if context:
        payload["context"] = context
    
    # Make the POST request to the API
    response = requests.post(url, headers=headers, json=payload)
    return response.json()
```

### Fact-checking

Let's test fact-checking mode with our examples:

```python
# Test fact-checking mode
fact_results = []

for i, example in enumerate(fact_checking_examples):
    print(f"=== Fact-checking Example {i+1} ===")
    print(f"Question: {example['prompt']}")
    print(f"Answer: {example['output']}")
    print(f"Expected Hallucination: {example['expected_hallucination']}")
    print()
    
    result = detect_hallucination_qa(
        prompt=example['prompt'],
        output=example['output'],
        return_search_results=False
    )
    
    fact_results.append({
        'example': example,
        'result': result
    })
    
    print("Detection Result:")
    print(f"Is Hallucination: {result.get('is_hallucination', 'N/A')}")
    print(f"Explanation: {result.get('explanation', 'N/A')}")
    print(f"Tokens Used: {result.get('usage', {})}")
    print("\n" + "="*80 + "\n")
```

### Enable search results

When enabling the property `return_search_results=true`, the agent will return the sources used for the fact-checking.

```python
# Test fact-checking mode with search results enabled
print("=== Fact-checking with Search Results ===")
example = {
    "prompt": "Are ghosts real?",
    "output": "Yes, there is a recent scientific study from Harvard that confirms ghosts exist."
}

# lets run the detection with search results enabled
result = detect_hallucination_qa(
    prompt=example['prompt'],
    output=example['output'],
    return_search_results=True # Enable search results 
)

# Display the result
print(f"Question: {example['prompt']}")
print(f"Answer: {example['output']}")
print(f"\nIs Hallucination: {result.get('is_hallucination', 'N/A')}")


# Display search results
if 'search_results' in result and result['search_results']:
    print(f"\nðŸ“š Search Results Used ({len(result['search_results'])} sources):")
    for idx, source in enumerate(result['search_results'][:5], 1):  # Show first 5
        print(f"\n{idx}. {source.get('title', 'No title')}")
        print(f"   ðŸ“„ {source.get('snippet', 'No snippet')[:150]}...")
        print(f"   ðŸ”— {source.get('link', 'No link')}")
else:
    print("\nNo search results returned")

print(f"\nTokens Used: {result.get('usage', {})}")
```

### Context validation mode

The context validation mode uses the `context` property as the ground truth. When enabled, the agent does not fact-check the answer using external knowledge; instead, it focuses on identifying hallucinations based solely on the information within the provided `context`.

```python
# Test context validation mode
context_results = []

# 
for i, example in enumerate(context_validation_examples):
    print(f"=== Context Validation Example {i+1} ===")
    print(f"Context: {example['context']}")
    print(f"Question: {example['prompt']}")
    print(f"Answer: {example['output']}")
    print(f"Expected Hallucination: {example['expected_hallucination']}")
    print()
    
    # Run the hallucination detection with context
    result = detect_hallucination_qa(
        prompt=example['prompt'],
        output=example['output'],
        context=example['context'],
        return_search_results=False
    )
    
    context_results.append({
        'example': example,
        'result': result
    })
    
    # Display the results
    print("Detection Result:")
    print(f"Is Hallucination: {result.get('is_hallucination', 'N/A')}")
    print(f"Explanation: {result.get('explanation', 'N/A')}")
    print(f"Tokens Used: {result.get('usage', {})}")
    print("\n" + "="*80 + "\n")
```

## Chat completion

The Chat Completion method validates multi-turn conversations for hallucinations. This is ideal for conversational AI systems and chatbots.

### Prepare chat data

```python
# Chat conversation examples for Chat Completion detection
chat_examples = [
    {
        "messages": [
            {"role": "system", "content": "You are a helpful assistant that provides accurate medical information."},
            {"role": "user", "content": "Does vitamin C cure the common cold?"},
            {"role": "assistant", "content": "Yes, taking large doses of vitamin C has been scientifically proven to cure the common cold within 24 hours."}
        ],
        "expected_hallucination": True
    },
    {
        "messages": [
            {"role": "system", "content": "You are a knowledgeable financial advisor."},
            {"role": "user", "content": "What is compound interest?"},
            {"role": "assistant", "content": "Compound interest is when you earn interest on both your original investment and the interest that has already been earned. It causes your money to grow exponentially over time."}
        ],
        "expected_hallucination": False
    }
]
```

### Use OpenAI Client

```python
# Function for Chat Completion hallucination detection
def detect_hallucination_chat(messages):
    """Detect hallucination in chat conversations using OpenAI SDK format"""
    
    # Create a separate client for chat completions with the correct base URL
    client = OpenAI(
        base_url=base_url,
        api_key=api_key,
    )
    
    # Make the request using OpenAI client - pass parameters directly
    response = client.chat.completions.create(
        model="klusterai/hallucination-detection",
        messages=messages
    )
    
    # Parse the response - kluster.ai returns detection results in a specific format
    return response.choices[0].message.content
```

```python
# Test Chat Completion detection with our examples
chat_results = []

for i, example in enumerate(chat_examples):
    print(f"=== Chat Example {i+1} ===")
    print(f"System: {example['messages'][0]['content']}")
    print(f"User: {example['messages'][1]['content']}")
    print(f"Assistant: {example['messages'][2]['content']}")
    print(f"Expected Hallucination: {example['expected_hallucination']}")
    print()
    
    try:
        result = detect_hallucination_chat(
            messages=example['messages'],
            
        )
        
        chat_results.append({
            'example': example,
            'result': result
        })
        
        print("Detection Result:")
        print(f"Result: {result}")
        
    except Exception as e:
        print(f"Error processing chat example: {e}")
    
    print("\n" + "="*80 + "\n")

# Summary of chat detection results
print("### Chat Detection Summary")
print(f"Processed {len(chat_results)} chat conversations")
for i, result in enumerate(chat_results):
    expected = result['example']['expected_hallucination']
    print(f"Chat {i+1}: Expected hallucination = {expected}")
```
